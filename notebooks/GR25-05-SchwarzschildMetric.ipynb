```python
M = Manifold(4, 'M', structure='Lorentzian')
U = M.open_subset('U')
X.<t,r,theta,phi>= U.chart(r"t r theta:(0,pi):\theta phi:(0,2*pi):\phi")
```

```python
A = function('A')
B = function('B')
```

```python
g = M.metric('g')
g[0,0] = -exp(2*A(r))
g[1,1] = exp(2*B(r))
g[2,2] = r^2
g[3,3] = r^2*sin(theta)^2
```

```python
g.display(X.frame())
```

```python
nabla = g.connection()
nabla.display(X.frame())
```

```python
Gamma = nabla.coef(X.frame())
show(Gamma)
```

```python
Riem = g.riemann()
Riem.display(X.frame())
```

```python
Ric = g.ricci()
Ric.display(X.frame())
```

```python
R = g.ricci_scalar()
R.display()
```

```python
G = Ric - (1/2)*R*g
G.set_name('G', latex_name=r'G')
G.display(X.frame())
```

```python
Gtt = G[0,0]
Grr= G[1, 1]
Gtht = G[3,3]
show(r'G_tt=', Gtt)
show(r'G_rr=', Grr)
show(r'G_theta_theta=',Gtht)
```

```python
eq = (exp(-2*A(r))*Gtt.expr() + exp(-2*B(r))*Grr.expr()).simplify_full()
show(eq)
```

```python
Grr_expr = Grr.expr() 
Grr_sub = Grr_expr.subs({B(r): -A(r), diff(B(r),r): -diff(A(r),r)}).simplify_full()
show(Grr_sub)
```
