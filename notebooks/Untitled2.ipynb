```python
M = Manifold(4, 'M', structure='Lorentzian')
U = M.open_subset('U')
X.<t, r, theta, phi>=U.chart(r"t r theta:(0,pi):\theta phi:(0,2*pi):\phi")
```

```python
A = function('A')
B = function('B')
```

```python
g = M.metric('g')
g[0,0]=-exp(2*A(r))
g[1,1]=exp(2*B(r))
g[2,2]=r^2
g[3,3]=r^2*sin(theta)^2
```

```python
g.display()
```

```python
nabla = g.connection()
nabla.display(X.frame())
```

```python
Riem = g.riemann()
Riem.display(X.frame())
```

```python
Ric = g.ricci()
Ric.display(X.frame())
```

```python
R = g.ricci_scalar()
R.display()
```

```python
G = Ric - (1/2)*R*g
G.set_name('G', latex_name='G')
G.display(X.frame())
```

```python
Gtt = G[0,0]
Grr = G[1,1]
Gtht = G[2,2]
```

```python
show('G_tt=', Gtt)
show('G_rr=', Grr)
show('G_thth=', Gtht)
```

```python
eq = (exp(-2*A(r))*Gtt.expr() + exp(-2*B(r))*Grr.expr()).simplify_full()
```

```python
show(eq)
```

```python
Grr_expr = Grr.expr() 
Grr_sub = Grr_expr.subs({B(r): -A(r), diff(B(r),r): -diff(A(r),r)}).simplify_full()
show(Grr_sub)
```

```python
M4 = Manifold(4, 'M4', structure='Lorentzian')
U = M4.open_subset('U')
X.<t,r,th,ph> = U.chart(r"t r th:(0,pi):\theta ph:(0,2*pi):\phi")

var('G M') 
f = 1 - 2*G*M/r
g = M4.metric('g')
g[0,0] = -f                    # g_tt  = -(1 - 2GM/r)
g[1,1] =  1/f                  # g_rr  = 1/(1 - 2GM/r)
g[2,2] =  r^2                  # g_θθ  = r^2
g[3,3] =  r^2*sin(th)^2     # g_φφ  = r^2 sin^2θ
Ric = g.ricci()
Rscalar = g.ricci_scalar()
Riem = g.riemann()

Ric_comp = Ric.comp(X.frame())
print("Ricci tensor components:")
for i in range(4):
    for j in range(4):
        print(f"Ric[{i},{j}] =", Ric_comp[i,j].expr().simplify_full())

Rscalar = g.ricci_scalar()
print("\nRicci scalar R:")
print("R =", Rscalar.expr().simplify_full().n()) 

print("\nRicci tensor displayed:")
Ric.display(X.frame())

Riem_comp = Riem.comp(X.frame())

print("\nSome Riemann components R^μ_{ νρσ}:")
for idx in [(0,1,0,1), (2,3,2,3)]:
    mu, nu, rho, sig = idx
    val = Riem_comp[mu,nu,rho,sig].expr().simplify_full()
    print(f"R^{mu}_{nu}{rho}{sig} =", val)
```
