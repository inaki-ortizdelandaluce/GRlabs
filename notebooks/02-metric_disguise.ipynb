### Exercise 1.5
Consider the following metric:
$$
dl^2 = du^2 + dv^2 + dw^2 - \left(\frac{3}{13}du + \frac{4}{13}dv + \frac{12}{13}dw\right)^2
$$

It appears that this is a three-dimensional metric in u, v, and w. This is not true! This
metric is in fact two-dimensional in disguise. Prove this by finding a transformation between
the “coordinates” u, v, and w and the Cartesian coordinates x and y, for
example, that reduces the metric to:

$$
dl^2 = dx^2 + dy^2
$$

#### Simplifly expression

```python
var('du dv dw')
dl2 = du^2 + dv^2 + dw^2 - (3/13*du + 4/13*dv + 12/13*dw)^2
dl2 = dl2.expand()
dl2
```

#### Extract coefficients and build metric matrix

```python
coeff_du2 = dl2.coefficient(du^2)
coeff_dv2 = dl2.coefficient(dv^2)
coeff_dw2 = dl2.coefficient(dw^2)
coeff_dudv = dl2.coefficient(du*dv)
coeff_dudw = dl2.coefficient(du*dw)
coeff_dvdw = dl2.coefficient(dv*dw)
```

```python
A = Matrix([
    [coeff_du2, coeff_dudv/2, coeff_dudw/2],
    [coeff_dudv/2, coeff_dv2, coeff_dvdw/2],
    [coeff_dudw/2, coeff_dvdw/2, coeff_dw2]
])
A
```

#### Diagonalize metric and check dimensions of new coordinate system
Confirm metric is two dimensional in disguise by checking eigenvalues such that

$$
dl^2 = dx^2 + dy^2
$$

```python
eigenvalues = A.eigenvalues()
eigenvectors = A.eigenvectors_right()
eigenvalues
```

#### Calculate change of basis matrix and coordinate system transformations

```python
Q = Matrix([eigenvectors[0][1][0], 
            eigenvectors[0][1][1], 
            eigenvectors[1][1][0]]).transpose() 
Q*Matrix([[1, 0, 0],[0, 1, 0],[0, 0, 0]])*Q.inverse() == A
Q.inverse()*A*Q
```

```python
var('u v w')
uvw = vector([u , v, w])
transform = Q.inverse() * uvw
transform
```
